```python
df['Income'] = df['Income'].fillna(df['Income'].median())
```

```python
df['Age'] = 2024 - df['Year_Birth']
```

### customer total spending

```python
df['Total_Spending'] = (
    df['MntWines']
    + df['MntFruits']
    + df['MntMeatProducts']
    + df['MntFishProducts']
    + df['MntSweetProducts']
    + df['MntGoldProds']
)
```

## customer total purchase

```python
df['Total_Purchases'] = (
    df['NumWebPurchases']
    + df['NumCatalogPurchases']
    + df['NumStorePurchases']
)
```

###  customer spending distribution

```python
sns.histplot(df['Total_Spending'], bins=50)
plt.title("Customer Spending Distribution")
```

### Income vs Spending

```python
sns.scatterplot(x='Income', y='Total_Spending', data=df)
```

### Campaign Response Rate

```python
df['Response'].value_counts()
```

### Visualization:

```python
sns.countplot(x='Response', data=df)
```

### Which Customers Responded?

```python
sns.boxplot(x='Response', y='Income', data=df)
```

### Purchase Channel Analysis

```python
channels = ['NumWebPurchases','NumCatalogPurchases','NumStorePurchases']

df[channels].sum().plot(kind='bar')
plt.title("Purchase Channel Distribution")
```

### Recency Analysis

```python
sns.boxplot(x='Response', y='Recency', data=df)
```

### Customer Segmentation

```python
df['Customer_Segment'] = pd.qcut(df['Total_Spending'], 3, labels=['Low','Medium','High'])
```

### visualization

```python
sns.countplot(x='Customer_Segment', data=df)
```
